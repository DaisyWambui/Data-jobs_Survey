<h2>Data cleaning</h2><br>
1.Drop Irrelevant columns<br>
2.Splitting the 'Q3 - Current Yearly Salary (in USD)' column into two and calculating the averages.<br>
3.Renaming the long columns to shorter names<br>
4.Mapping various job titles to a standardized title based on specific keywords present in the job titles.<br>
5.Mapping various Industries to a standardized title based on specific keywords present in the industries.<br>
6.Mapping programming languages to a standardized title based on specific keywords present in the languages.<br>
7.Consolidating Desired job factors.<br>
8.Consolidating countries into continents.<br>

```python
#Dropping irrelevant columns

tech_jobs.drop(['Browser', 'OS', 'Email', 'City', 'Country', 'Referrer','Q13 - Ethnicity', 'Date Taken (America/New_York)','Time Taken (America/New_York)', 'Time Spent'], axis=1, inplace=True)
```

```python
tech_jobs.shape
```

```python
#Splitting the 'Q3 - Current Yearly Salary (in USD)' column into two

tech_jobs[['Salary1', 'Salary2']] =tech_jobs['Q3 - Current Yearly Salary (in USD)'].str.split('-', expand=True)
```

```python
tech_jobs[['Salary1','Salary2']]
```

```python
#Removing the letters in the salary column

tech_jobs['Salary1(USD)'] = tech_jobs['Salary1'].str.replace('k', '').str.replace('+', '').astype(float) * 1000
tech_jobs['Salary2(USD)'] = tech_jobs['Salary2'].str.replace('k', '').astype(float) * 1000
```

```python
tech_jobs[['Salary1(USD)','Salary2(USD)']]
```

```python
#Creating a new column which contains averages of the salary column

tech_jobs['AverageSalary'] = (tech_jobs['Salary1(USD)'] + tech_jobs['Salary2(USD)']) / 2
tech_jobs['AverageSalary']
```

```python
#Dropping irrelevant columns

tech_jobs.drop(['Salary1','Salary2', 'Salary1(USD)','Salary2(USD)','Q3 - Current Yearly Salary (in USD)'], axis=1, inplace=True)
tech_jobs.head()
```

```python
#Renaming the long columns to shorter names

tech_jobs.rename(columns={'Q1 - Which Title Best Fits your Current Role?': 'Current Role Title', 
                          'Q2 - Did you switch careers into Data?': 'Switched to Data',
                          'Q4 - What Industry do you work in?':'Industry',
                          'Q5 - Favorite Programming Language':'ProgLanguage',
                          'Q6 - How Happy are you in your Current Position with the following? (Salary)':'Current Happiness (Salary)',
                          'Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)':'Current Happiness (Work/Life)',
                          'Q6 - How Happy are you in your Current Position with the following? (Coworkers)':'Current Position Happiness (Coworkers)',
                          'Q6 - How Happy are you in your Current Position with the following? (Management)':'Current Position Happiness (Management)',
                          'Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)':'Current Position Satisfaction (Upward Mobility)',
                          'Q6 - How Happy are you in your Current Position with the following? (Learning New Things)':'Learning Happiness',
                          'Q7 - How difficult was it for you to break into Data?':'Data Break-In',
                          'Q8 - If you were to look for a new job today, what would be the most important thing to you?':'Desired Job Factor',
                          'Q9 - Male/Female?':'Gender','Q10 - Current Age':'Current Age','Q11 - Which Country do you live in?':'Residence',
                          'Q12 - Highest Level of Education':'Highest Level of Education','Q13 - Ethnicity':'Ethnicity'}, inplace=True)
tech_jobs.head()
```

<h3>Consolidating countries to continents.<h3/>

```python
tech_jobs['Residence'].unique()
```

```python
african_countries = ['Algeria', 'Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 'Cabo Verde', 'Cameroon', 
                     'Central African Republic', 'Chad', 'Comoros', 'Congo', 'Cote d\'Ivoire', 'Djibouti', 'Egypt', 
                     'Equatorial Guinea', 'Eritrea', 'Eswatini', 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 
                     'Guinea-Bissau', 'Kenya', 'Lesotho', 'Liberia', 'Libya', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 
                     'Mauritius', 'Morocco', 'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Sao Tome and Principe',
                     'Senegal', 'Seychelles', 'Sierra Leone', 'Somalia', 'South Africa', 'South Sudan', 'Sudan', 'Tanzania',
                     'Togo', 'Tunisia', 'Uganda', 'Zambia', 'Zimbabwe']

tech_jobs['Residence'] = tech_jobs['Residence'].apply(lambda x: 'Africa' if x in african_countries else x)
```

```python
tech_jobs['Residence'] = [role.split(':')[-1].strip() for role in tech_jobs['Residence']]
tech_jobs['Residence'].unique()
```

```python
countries_to_change = ['Kenya', 'Nigeria', 'South Africa','nigeria', 'Republic democratic of Congo ','Somalia','Ghana',
                       'Kenua','Tunisia','Niger','Republic democratic of Congo', 'Sudan','Kenya ','Nigeria ','Egypt',
                       'Mozambique',' Somalia','Morocco','Africa (Nigeria)','Angola','Liberia','Algeria']  


# Update the values in the "residence" column
tech_jobs.loc[tech_jobs['Residence'].isin(countries_to_change), 'Residence'] = 'Africa'
```

```python

# Create a mapping dictionary of countries and their corresponding continents
country_to_continent = {
    'United States': 'North America',
    'Canada': 'North America',
    'Africa': 'Africa',
    'Serbia': 'Europe',
    'Argentina': 'South America',
    'United Kingdom': 'Europe',
    'Azerbaijan': 'Asia',
    'India': 'Asia',
    'Japan': 'Asia',
    'Greece': 'Europe',
    'Perú': 'South America',
    'Australia': 'Oceania',
    'Spain': 'Europe',
    'Barbados': 'North America',
    'Colombia': 'South America',
    
}

# Add a new column "continent" based on the mapping
tech_jobs['Residence'] = tech_jobs['Residence'].map(country_to_continent)


```
